```python
import pandas as pd
import os
import glob

base_path = "ë°ì´í„°"

# ëª¨ë“  í•˜ìœ„ í´ë” í¬í•¨ CSV íŒŒì¼ ê²€ìƒ‰
csv_files = glob.glob(os.path.join(base_path, "*", "*.csv"))

for file in csv_files:
    # ê²½ë¡œ ë¶„ë¦¬
    parts = file.replace("\\", "/").split("/")
    folder = parts[-2]
    filename = parts[-1].replace(".csv", "")
    
    # ì €ì¥ ê²½ë¡œ ì§€ì • (ê°™ì€ í´ë” ë‚´ ë™ì¼ ì´ë¦„ìœ¼ë¡œ ì €ì¥)
    save_path = os.path.join(base_path, folder, f"{filename}.xlsx")

    try:
        # CSV ë¶ˆëŸ¬ì˜¤ê¸° (utf-8 ìš°ì„ , ì‹¤íŒ¨ ì‹œ cp949 ì¬ì‹œë„)
        try:
            df = pd.read_csv(file, encoding="utf-8", on_bad_lines="skip")
        except UnicodeDecodeError:
            df = pd.read_csv(file, encoding="cp949", on_bad_lines="skip")
        
        # Excelë¡œ ì €ì¥
        df.to_excel(save_path, index=False, engine="openpyxl")
        print(f"âœ… ë³€í™˜ ì™„ë£Œ: {save_path} ({df.shape[0]}í–‰, {df.shape[1]}ì—´)")

    except Exception as e:
        print(f"âš ï¸ ë³€í™˜ ì‹¤íŒ¨: {file} â†’ {e}")

print("\nğŸ“¦ ëª¨ë“  CSV â†’ Excel ë³€í™˜ ì™„ë£Œ")
```

```python
import pandas as pd

# 1ï¸âƒ£ ë¸”ë¡œê·¸ ë°ì´í„° ì „ì²˜ë¦¬
df_blog = df.copy()
df_blog = df_blog.rename(columns={
    "webpage_url": "url",
    "post_wrtdt": "date"
})
df_blog["channel"] = "blog"
if "date" not in df_blog.columns:
    df_blog["date"] = None

# 2ï¸âƒ£ íŠ¸ìœ„í„° ë°ì´í„° ì „ì²˜ë¦¬
df_twitter = df1.copy()
df_twitter = df_twitter.rename(columns={
    "tweet_url": "url"
})
df_twitter["channel"] = "twitter"
if "title" not in df_twitter.columns:
    df_twitter["title"] = None
if "date" not in df_twitter.columns:
    df_twitter["date"] = None

# 3ï¸âƒ£ ê³µí†µ ì»¬ëŸ¼ ë§ì¶”ê¸° (ì—†ìœ¼ë©´ ìƒì„±)
# ì´ë¯¸ ìœ„ì—ì„œ ìƒì„±í•¨

# 4ï¸âƒ£ ì»¬ëŸ¼ ìˆœì„œ í†µì¼
for col in ["url", "title", "text", "date", "channel"]:
    if col not in df_blog.columns:
        df_blog[col] = None
    if col not in df_twitter.columns:
        df_twitter[col] = None
df_blog = df_blog[["url", "title", "text", "date", "channel"]]
df_twitter = df_twitter[["url", "title", "text", "date", "channel"]]

# 5ï¸âƒ£ ë³‘í•©
df_merged = pd.concat([df_blog, df_twitter], ignore_index=True)

# 6ï¸âƒ£ ê²°ê³¼ í™•ì¸
print(df_merged.info())
print(df_merged.head())
```

```python
import pandas as pd

# ë¨¼ì € dateê°€ ê²°ì¸¡ì¹˜ì¸ í–‰ì„ ì œê±°í•©ë‹ˆë‹¤.
df_merged = df_merged.dropna(subset=["date"])

# date ì»¬ëŸ¼ì„ datetimeìœ¼ë¡œ ë³€í™˜ ê°€ëŠ¥í•œ í–‰ë§Œ ë‚¨ê¹ë‹ˆë‹¤.
def is_datetime(val):
    try:
        pd.to_datetime(val)
        return True
    except Exception:
        return False

df_merged = df_merged[df_merged["date"].apply(is_datetime)]
df_merged.head()
```

```python
import os
import pandas as pd

# ë°ì´í„° í´ë” ë‚´ ê° ì´ìŠˆë³„ ë³‘í•© ì—‘ì…€ íŒŒì¼ì—ì„œ ë°ì´í„° ê°œìˆ˜ë¥¼ ì¹´ìš´íŠ¸í•˜ëŠ” ìš”ì•½ í…Œì´ë¸” ì¶œë ¥
data_folder = "ë°ì´í„°"
excel_files = [
    "merged_output_gs.xlsx",
    "merged_output_KT.xlsx",
    "merged_output_lotte.xlsx",
    "merged_output_SKT.xlsx"
]

summary = []

for file in excel_files:
    file_path = os.path.join(data_folder, file)
    issue_name = file.replace("merged_output_", "").replace(".xlsx", "")
    if os.path.exists(file_path):
        df = pd.read_excel(file_path)
        count = len(df)
    else:
        count = 0
    summary.append({"ì´ìŠˆ": issue_name, "ë°ì´í„°ìˆ˜": count})

summary_df = pd.DataFrame(summary)
print("ì´ìŠˆë³„ ë°ì´í„° ê°œìˆ˜ ìš”ì•½:")
print(summary_df)

```
